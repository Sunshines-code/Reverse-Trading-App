```python
import pandas as pd
from pprint import pprint
#This file imports data for all tradable stocks on the Alpacca network, sorts them based on average change in stock
#price and volume before saving the sections to make machine learning models.
```

```python
#The csv file inported holds 400 days of data for all us traded stocks
data = pd.read_csv('./total_us_stock_data400_daysccV.csv')
total_df = pd.DataFrame(data)
total_df
```

```python
#The list imported here has only the stocks that Alpacca states they trade (discrepencies are occasionally found)
data = pd.read_csv('./Alpacca_NYSE_NASDAQ_List.csv')
tckr = data['Tckr'].tolist()
tckr
```

```python
tradeablu_huge_df = total_df[total_df.Ticker.isin(tckr)]
```

```python
#creating columns for daily price change and volume exchanged. I did some reading after this file was made and 
#stopped using volume as an indicator for price predicting. However when exiting a stock the speed of doing so can 
#be fairly well guessed based on the volume traded per day.
tradeablu_huge_df['change']= (abs(tradeablu_huge_df['high'])/tradeablu_huge_df['low']*100)-100
tradeablu_huge_df['DV']= (abs(tradeablu_huge_df['vol'])*tradeablu_huge_df['low']*100)-100

tradeablu_huge_df
```

```python
TckrDf = tradeablu_huge_df.groupby(tradeablu_huge_df['Ticker'])
```

```python
#reviewing this I am not sure why i didnt drop the unneeded columns but I get a new df eventually
changeList = TckrDf['change'].mean()
volList = TckrDf['vol'].mean()

changeList
```

```python
Changedf = pd.DataFrame(changeList)
Changedf['Av'] = volList
Changedf
```

```python
#Now that I have the new df I am sorting by daily volume so as to drop the stocks that are rarely traded
#The first iteration of this process involved the top traded 5,000 but after a lot of issues with slippage 
#I changed it to the top 2000. This changed the average amount of chaange but essentially eleminated slippage.
Changedf = Changedf.sort_values(by='Av', ascending=False)
# Changedf = Changedf.dropna()
Changedf = Changedf.reset_index(drop=False)
# gen5df = Changedf.loc[800:1471]
gen6df = Changedf.head(2001)
gen6df
```

```python
#Resorting by change
#That is the variable I was looking to explore for a zone of balance between predicability and profit
gen6df= gen6df.sort_values(by='change', ascending=False)
gen6df = gen6df.drop(0, axis=0)
gen6df = gen6df.reset_index(drop=True)
gen6df
```
