```python
data2 = []
for t in save_list:
    model = load(f'./v3100/{t}Model.joblib')
    X_scaler = load(f'./v3100/{t}X_scaler.joblib') 
    y_scaler = load(f'./v3100/{t}y_scaler.joblib')
    df = pd.read_csv(f'.//v3100/{t}StockListMAData.csv')
    df = pd.DataFrame(df)
    df = df.sort_values(by='day', ascending=True)
    df =df.reset_index(drop=True)
    q=0
    c =0
    change = 0
    NDays = 0
    tchange = 0

    for i in range(1,90):
        try:
            b = (i*50 - 50)+c
            e = (i*49)+c
            cdf = df.loc[b:e]
            ldf = cdf[['opening', 'closing', 'high', 'low']]
            sc_new_in = X_scaler.transform(ldf)
            new_out = model.predict(sc_new_in)
            new_out = y_scaler.inverse_transform(new_out)
            cdf["next_high"]= new_out
            cdf["change"]= cdf['next_high']/cdf["closing"]*100-100
            rdf = cdf.sort_values(by=['change'], axis=0, ignore_index=True, inplace=False, ascending = False)
            actHigh = rdf.loc[0,].NextHigh
            closing = rdf.loc[0,].closing
            if actHigh >= closing:
                dif = actHigh/closing*100-100
                change = change +dif
                tchange = tchange + dif
            else:
                dif = actHigh/closing*100-100
                change-=1
                NDays +=1
                tchange = tchange + dif
#             print(dif)

            c+=1
        except:
            print(i)
            pass
    info = {"model": f"{t}model",
           'increase':change,
           "Ndays":NDays,
           "ActChange":tchange}
    print(save_list[q])
    data2.append(info)
    
    q+=1
```
