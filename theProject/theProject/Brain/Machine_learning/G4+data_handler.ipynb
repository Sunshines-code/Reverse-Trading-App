```python
import pandas as pd
from pprint import pprint
```

```python
data = pd.read_csv('./total_us_stock_data400_days.csv')
total_df = pd.DataFrame(data)
total_df
```

```python
data = pd.read_csv('../Machine_learning/Alpacca_NYSE_NASDAQ_List.csv')
tckr = data['Tckr'].tolist()
tckr
```

```python
tradeablu_huge_df = total_df[total_df.Ticker.isin(tckr)]
```

```python
tradeablu_huge_df['change']= (abs(tradeablu_huge_df['high'])/tradeablu_huge_df['low']*100)-100
tradeablu_huge_df
```

```python
TckrDf = tradeablu_huge_df.groupby(tradeablu_huge_df['Ticker'])
```

```python
changeList = TckrDf['change'].mean()
changeList
```

```python
Changedf = pd.DataFrame(changeList)
Changedf = Changedf.sort_values(by='change', ascending=False)
# gen5df = gen5df.drop([0, 50])
# gen5df = gen5df.head(500)
Changedf = Changedf.dropna()
Changedf = Changedf.reset_index(drop=False)
gen5df = Changedf.loc[4000:4500]
gen5df
```

```python
gen5df = gen5df.reset_index(drop=True)
gen5df
```

```python
gen5TckrList = gen5df['Ticker'].tolist()
gen5Tckr = {'Tckr': gen5TckrList}
gen5Tckr = pd.DataFrame(gen5Tckr)
gen5Tckr.to_csv(r'../Machine_Learning/G12List.csv', index = False)
gen5Tckr
```

```python
G5trainingData = total_df[total_df.Ticker.isin(gen5TckrList)]
G5trainingData = G5trainingData.reset_index(drop=True)
G5trainingData.to_csv(r'../Machine_Learning/MLData12.csv', index = False)
G5trainingData
```
