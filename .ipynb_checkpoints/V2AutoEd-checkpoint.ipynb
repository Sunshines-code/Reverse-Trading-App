```python
# This file educates models unique for each stock list created in the sort and seperate file.
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, BayesianRidge, Lasso, Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import json
import pprint as pprint
import pandas as pd
from joblib import dump, load
file_list=['V3_1StockList', 'V3_2StockList', 'V3_3StockList', 'V3_4StockList', 'V3_5StockList', 'V3_6StockList', 'V3_7StockList', 'V3_8StockList', 'V3_9StockList', 'V3_10StockList', 'V3_11StockList', 'V3_12StockList', 'V3_13StockList', 'V3_14StockList', 'V3_15StockList', 'V3_16StockList', 'V3_17StockList', 'V3_18StockList', 'V3_19StockList', 'V3_20StockList', 'V3_21StockList', 'V3_22StockList', 'V3_23StockList', 'V3_24StockList', 'V3_25StockList', 'V3_26StockList', 'V3_27StockList', 'V3_28StockList', 'V3_29StockList', 'V3_30StockList', 'V3_31StockList', 'V3_32StockList', 'V3_33StockList', 'V3_34StockList', 'V3_35StockList', 'V3_36StockList', 'V3_37StockList', 'V3_38StockList', 'V3_39StockList', 'V3_40StockList']
```

```python
label_list=[]
for i in range(1,41):
    label_list.append(f"V3_{i}")
print(label_list)
```

```python
#This loops through educating each new model using information from the ModelEdDataCall file
#After educating and storing the needed files in the V3100 foulder it gets and MSE and R2 score as well. 
c=0
comp_list = []
for i in file_list:
    print(f'./{label_list[c]}')
    df = pd.read_csv(f'./v3100/{i}edData.csv')
    df = pd.DataFrame(df)
    X = df[['opening', 'closing', 'high', 'low']]
    y = df["NextHigh"].values.reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    X_scaler = StandardScaler().fit(X_train)
    y_scaler = StandardScaler().fit(y_train)
    dump(X_scaler, f'./v3100/{label_list[c]}X_scaler.joblib') 
    dump(y_scaler, f'./v3100/{label_list[c]}Y_scaler.joblib')
    model = LinearRegression()
    X_train_scaled = X_scaler.transform(X_train)
    X_test_scaled = X_scaler.transform(X_test)
    y_train_scaled = y_scaler.transform(y_train)
    y_test_scaled = y_scaler.transform(y_test)
    model.fit(X_train_scaled, y_train_scaled)
    predictions = model.predict(X_test_scaled)
    MSE = mean_squared_error(y_test_scaled, predictions)
    r2 = model.score(X_test_scaled, y_test_scaled)
    dump(model, f'./v3100/{label_list[c]}model.joblib') 
    print(f"MSE: {MSE}, R2: {r2}")
    data= {"model": f'{label_list[c]}',
          "MSE":MSE, "R2":r2}
    comp_list.append(data)

    print(label_list[c])

    c+=1

#     break
    

```

```python
# I collected this data to see how strong of a correlation between change % per day was seen
#in a specific model/stock group and the R2 value. A trend of increased accuracy can be seen in the list below with outlyers throughout
tdf = pd.DataFrame(comp_list)
tdf.to_csv(f"./V3100ModelInfoData.csv", index = False)
tdf
```
