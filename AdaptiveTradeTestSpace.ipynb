```python
#this file is for creating the adaptive trading section and running it through 90 days of simulated trading. 
#Adaptive meaning that the model applied changes if a model cannot reach an acceptable R2 score then the next models will be given a chance.
#this will hopefully decrease negative days of trading and possibly maintain positive percent change. 
#Success would look like <10 bad days of trading with >2,000% change.
#create new simulater-run small cluster, add in outliers, then try top 20 ranked by proffit also possibly let all models be tested in order of ActChange
from sklearn.linear_model import LinearRegression, BayesianRidge, Lasso, Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import json
import pprint as pprint
import pandas as pd
from joblib import dump, load
```

```python
label_list= ['V4_164', 'V4_165', 'V4_166', 'V4_167', 'V4_168','V4_169', 'V4_170', 'V4_171', 'V4_172']  
```

```python
data2 = []
change = 0
NDays = 0
tchange = 0

for i in range(0,89):
#     try:
    b = (i*50)
    e = (i*50+49)
    print(i+1)
    for t in label_list:
        print(t)
        model = load(f'./V4-2000/{t}Model.joblib')
        X_scaler = load(f'./V4-2000/{t}X_scaler.joblib') 
        y_scaler = load(f'./V4-2000/{t}y_scaler.joblib')
        df = pd.read_csv(f'./V4-2000/{t}MAData.csv')
        df = pd.DataFrame(df)
        df = df.sort_values(by='day', ascending=True)
        df =df.reset_index(drop=True)
        cdf = df.loc[b:e]
        print(cdf.head())
        X = df[['PrevOpening', 'PrevClosing', 'PrevHigh', 'PrevLow']]
        y = df["high"].values.reshape(-1, 1)
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
        model = LinearRegression()
        X_train_scaled = X_scaler.transform(X_train)
        X_test_scaled = X_scaler.transform(X_test)
        y_train_scaled = y_scaler.transform(y_train)
        y_test_scaled = y_scaler.transform(y_test)
        model.fit(X_train_scaled, y_train_scaled)
        predictions = model.predict(X_test_scaled)
        r2 = model.score(X_test_scaled, y_test_scaled)
        
        ldf = cdf[['opening', 'closing', 'high', 'low']]
        sc_new_in = X_scaler.transform(ldf)
        new_out = model.predict(sc_new_in)
        new_out = y_scaler.inverse_transform(new_out)
        cdf["next_high"]= new_out
        cdf["change"]= cdf['next_high']/cdf["closing"]*100-100
        rdf = cdf.sort_values(by=['change'], axis=0, ignore_index=True, inplace=False, ascending = False)
        actHigh = rdf.loc[0,].NextHigh
        closing = rdf.loc[0,].closing
        if r2 > 0.9983:

            if actHigh >= closing:
                dif = actHigh/closing*100-100
                change = change +dif
                tchange = tchange + dif
                print(f'positive change {t} R2 {r2}')
            else:
                dif = actHigh/closing*100-100
                change-=1.5
                NDays +=1
                tchange = tchange + dif
                print(f'negative change {t} R2 {r2}')
            break

        else:
            print(f'skip {t} R2 {r2}')
            pass
            
#     except:
#         pass
    #         info = {"model": f"{t}model",
    #                'increase':change,
    #                "Ndays":NDays,
    #                "ActChange":tchange}
    #         print(t)
    #         data2.append(info)

```
