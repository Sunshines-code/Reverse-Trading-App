```python
for t in label_list:
    df = pd.read_csv(f"./V4-2000/{t}.csv")
    bdf = pd.DataFrame(df)
    tckr = bdf['Tckr'].tolist()
    price_info= []
    count = 0
    d = 0
    for i in tckr:

        try:
            barset = api.get_bars(i, TimeFrame.Day, "2021-11-04","2022-03-16", adjustment='raw')
            #Alpaca API V2 requires dates.
            d=0
            for f in range(0,90):
                tckr_infopopening = barset[d].o
                tckr_infopClosing = barset[d].c
                tckr_infopHigh = barset[d].h
                tckr_infopLow = barset[d].l
                d+=1
                tckr_infoopening = barset[d].o
                tckr_infoClosing = barset[d].c
                tckr_infoHigh = barset[d].h
                tckr_infoLow = barset[d].l 
                day = d

                tckr_infoNHigh = barset[d+1].h


                info = {'Ticker':f'{i}',
                        'opening':tckr_infoopening,
                        'closing':tckr_infoClosing,
                        'NextHigh':tckr_infoNHigh,
                        'high':tckr_infoHigh,
                        'low':tckr_infoLow,
                       'day':day,
                        'PrevOpening':tckr_infopopening,
                        'PrevClosing':tckr_infopClosing,
                        'PrevHigh':tckr_infopHigh,
                        'PrevLow':tckr_infopLow,                          }
                price_info.append(info)
#                 print(info)

        except:
            pass

        count+=1

#     break
    fdf = pd.DataFrame(price_info)
    fdf.to_csv(f"./V4-2000/{t}MAData.csv", index = False)
    print(t)
```
