```python
#This loops through educating each new model using information from the ModelEdDataCall file
#After educating and storing the needed files in the V3100 foulder it gets and MSE and R2 score as well. 
c=0
comp_list = []
for i in label_list:
    print(f'./{label_list[c]}')
    df = pd.read_csv(f'./V4-2000/{i}edData.csv')
    df = pd.DataFrame(df)
    X = df[['opening', 'closing', 'high', 'low']]
    y = df["NextHigh"].values.reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    X_scaler = StandardScaler().fit(X_train)
    y_scaler = StandardScaler().fit(y_train)
    dump(X_scaler, f'./V4-2000/{label_list[c]}X_scaler.joblib') 
    dump(y_scaler, f'./V4-2000/{label_list[c]}Y_scaler.joblib')
    model = LinearRegression()
    X_train_scaled = X_scaler.transform(X_train)
    X_test_scaled = X_scaler.transform(X_test)
    y_train_scaled = y_scaler.transform(y_train)
    y_test_scaled = y_scaler.transform(y_test)
    model.fit(X_train_scaled, y_train_scaled)
    predictions = model.predict(X_test_scaled)
    MSE = mean_squared_error(y_test_scaled, predictions)
    r2 = model.score(X_test_scaled,y_test_scaled)
    dump(model, f'./V4-2000/{label_list[c]}model.joblib') 
    print(f"MSE: {MSE}, R2: {r2}")
    data= {"model": f'{label_list[c]}',
          "MSE":MSE, "R2":r2}
    comp_list.append(data)

    print(label_list[c])

    c+=1

#     break
    

```

```python
# I collected this data to see how strong of a correlation between change % per day was seen
#in a specific model/stock group and the R2 value. A trend of increased accuracy can be seen in the list below with outlyers throughout
tdf = pd.DataFrame(comp_list)
tdf.to_csv(f"./V4ModelInfoData.csv", index = False)
tdf
```
